```octave
function p = predict(theta, X)
    m = size(X, 1);
    p = zeros(m, 1);
    hx = X*theta;
    for i =1:m
        if sigmoid(hx(i))>= 0.5,
            p(i) = 1;
        else
            p(i) = 0;
        end;
    end;
end
```

```octave
function p = predictOneVsAll(all_theta, X)
m = size(X, 1);
num_labels = size(all_theta, 1);
p = zeros(size(X, 1), 1);
X = [ones(m, 1) X]; % Add ones to the X data matrix
probMatrix = X*all_theta';
[pVector,p] = max(probMatrix,[],2);
end
```

```octave
function [all_theta] = oneVsAll(X, y, num_labels, lambda)
m = size(X, 1);
n = size(X, 2);
all_theta = zeros(num_labels, n + 1);
X = [ones(m, 1) X];
initial_theta = zeros(n+1,1);
options = optimset('GradObj', 'on', 'MaxIter', 50);
for(i=1:num_labels)
	y_b=(y==i);
	all_theta(i,:) = fmincg (@(t)(logisticRegressionCostFunction(X, y_b, t, lambda)), initial_theta, options);
end
end
```
