#### =========== Part 3: Train Linear Regression =============

```octave
%  Train linear regression with lambda = 0
lambda = 0;
initial_theta = zeros(size(X_bias, 2), 1); 
costFunction = @(t) linearRegressionCostFunction(X_bias, y, t, lambda);
options = optimset('MaxIter', 200, 'GradObj', 'on');
theta_1 = fmincg(costFunction, initial_theta, options);
```

```octave
p = 8;
%[X_poly, X_poly_cv, X_poly_test, mu, sigma] = polyFeaturesFinalTestNeedModify(X, Xcv, Xtest, p);
% Map X onto Polynomial Features and Normalize
X_poly = polynomialFeaturesMap(X, p);
% Normalizes the features in X 
mu = mean(X_poly);
sigma = std(X_poly);
X_poly = bsxfun(@rdivide, bsxfun(@minus, X_poly, mu), sigma);
X_poly = [ones(size(X_poly, 1), 1), X_poly];                   % Add Ones
% Map X_poly_val and normalize (using mu and sigma)
X_poly_cv = polynomialFeaturesMap(Xcv, p);
X_poly_cv = bsxfun(@rdivide, bsxfun(@minus, X_poly_cv, mu), sigma);
X_poly_cv = [ones(size(X_poly_cv, 1), 1), X_poly_cv]; % Add Ones
% Map X_poly_test and normalize (using mu and sigma)
X_poly_test = polynomialFeaturesMap(Xtest, p);
X_poly_test = bsxfun(@rdivide, bsxfun(@minus, X_poly_test, mu), sigma);
X_poly_test = [ones(size(X_poly_test, 1), 1), X_poly_test]; % Add Ones

lambda = 0;
initial_theta = zeros(size(X_poly, 2), 1); 
costFunction = @(t) linearRegressionCostFunction(X_poly, y, t, lambda);
options = optimset('MaxIter', 200, 'GradObj', 'on');
theta_8 = fmincg(costFunction, initial_theta, options);

lambda = 1;
initial_theta = zeros(size(X_poly, 2), 1); 
costFunction = @(t) linearRegressionCostFunction(X_poly, y, t, lambda);
options = optimset('MaxIter', 200, 'GradObj', 'on');
theta_f = fmincg(costFunction, initial_theta, options);
```

```octave
lambda = 0;
costFunction = @(a, b, c, d) linearRegressionCostFunction(a, b, c, d); % short hand
[error_train_1, error_cv_1, error_test_1] = learningCurve(costFunction, X_bias, y, Xcv_bias, ycv, Xtest_bias, ytest, lambda);
[error_train_8, error_cv_8, error_test_8] = learningCurve(costFunction, X_poly, y, X_poly_cv, ycv, X_poly_test, ytest, lambda);
[error_train_f, error_cv_f, error_test_f] = learningCurve(costFunction, X_poly, y, X_poly_cv, ycv, X_poly_test, ytest, 1);
```
